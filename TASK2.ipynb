```python
position_columns = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']
```

```python
position_df=new_df[position_columns]
position_df.isnull().sum()
```

```python
def convert_to_int(value):
    if isinstance(value, str) and '+' in value:
        return int(value.split('+')[0])
    return value


columns_to_convert = position_columns
for column in columns_to_convert:
    position_df[column] = position_df[column].apply(convert_to_int)

```

```python
avg_overall_by_position = position_df[position_columns].mean().reset_index()
avg_overall_by_position.columns = ['Position', 'Average Rating']
plt.figure(figsize=(12, 6))
sns.barplot(x='Position', y='Average Rating', data=avg_overall_by_position, palette='coolwarm')
plt.title("Average Overall Score by Position")
plt.xlabel("Position")
plt.ylabel("Average Overall Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
num_columns = len(position_columns)
cols = 3  # Number of columns in the grid
rows = (num_columns // cols) + (num_columns % cols > 0)  # Calculate number of rows needed

plt.figure(figsize=(15, rows * 5))  # Adjust height based on number of rows

# Loop through each position column and create a distribution plot
for i, column in enumerate(position_columns, 1):
    plt.subplot(rows, cols, i)  # Create subplot in the calculated grid
    sns.histplot(position_df[column], bins=5, kde=True, color='blue', stat='density')
    plt.title(f'Distribution of {column}')
    plt.xlabel('Rating')
    plt.ylabel('Density')

# Adjust layout
plt.tight_layout()
plt.show()
```
